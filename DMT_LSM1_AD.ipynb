```python
df=pd.read_csv("dataset_mood_smartphone.csv")
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
df=df.sort_values(by=['id', 'time'])
df
```

```python
df['time']=pd.to_datetime(df['time'])
```

```python
d=dict(tuple(df.groupby('id')))
print(d.keys())
d['AS14.33']
```

```python
data={}
for key in d.keys():
    dv=dict(tuple(d[key].groupby('variable')))
    # print(dv)
    df2=dv['mood'][["time","value"]]
    # print(df2)
    df2['time']=df2['time'].map(lambda x: x+datetime.timedelta(days=-5))
    # print(df2)
    df2=df2.set_index('time').groupby(pd.Grouper(freq='1D')).mean()
    # print(df2)
    # print("------------")
    for key2 in dv.keys():
        if key2 != 'mood':
            df3=dv[key2][["time","value"]].set_index('time').groupby(pd.Grouper(freq='1D')).mean()
            df3=df3.rolling(window='5D').mean().rename(columns={'value':key2})
            df2=df2.join(df3)
    data[key]=df2
data['AS14.02']
# the result of pre-processing is to summary the condition of following five days
# for example the fourth row in data['AS14.02'] means
# from 2014-3-14 to 2014-3-18, the average value of appCat.communication is 16.53
# and the value of 2014-3-19's mood is 6.8
# this is a very begining pre-processing
# we can polish together for various model and design different formats!!!
```
